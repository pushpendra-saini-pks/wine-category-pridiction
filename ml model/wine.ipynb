## Wine Category pridiction using KNN

```python
import numpy as np 
import pandas as pd 
```

```python
from sklearn.datasets import load_wine  
dataset = load_wine()
```

```python
dataset.keys()
```

```python
df = pd.DataFrame(dataset.data , columns=dataset.feature_names)
df.head()
```

```python
df['target'] = dataset.target
```

```python
x=dataset.data
y=dataset.target
```

```python
x,y
```

```python
Y = y.reshape(-1,1)
```

```python
Y.shape , x.shape
```

```python
type(Y)
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.20,random_state = 42)
```

```python
x_train 
```

```python
y_train 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors = 3 , algorithm = 'auto') 
```

```python
model 
```

```python
model.fit(x_train_scaled ,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
type(x_test),x_test.shape
```

```python
model.predict(np.array([13.16,	2.36,	2.67,	18.6,	101.0,	2.80,	3.24,	0.30,	2.81,	5.68,	1.03,	3.17,	1185.0]).reshape(1,-1))
```

```python
from sklearn.metrics import accuracy_score,classification_report
auc = accuracy_score(y_test,y_pred)
cls = classification_report(y_test , y_pred)
print(f"accuracy : {auc}")
print(f"classification_report : {cls}")
```

```python
from sklearn.metrics import precision_score , recall_score
ps = precision_score(y_test , y_pred,average="macro")
rs = recall_score(y_test , y_pred,average="macro")
```

```python
print(f"pricisin score : {ps}")
print(f"recall score : {rs}")
```
